## Compute lens curvature to inverse the sinusoidal trajectory of a resonant scanner

See blog post on ptrrupprecht.wordpress.com for more information.

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define angular range
alpha = np.arange(-np.pi/2+0.001,np.pi/2-0.001,0.01)
```

```python
# d distance between resonant scanner and lens; x distance from center of the lens for beam of angle alpha
d = 0.05
x = np.tan(alpha)*d
```

```python
# f(x) 'focal length', i.e., local inverse curvature of the lens to compensate for the sinusoidal trajectory
fx = x/(np.arctan(x/d) - np.arcsin(np.arctan(x/d)))
```

```python
# Plotting f(x)

# f(x) diverges for x = 0, indicating negligible local refractive power
# f(x) tends to zero for larger values of x, indicating high local refractive power

plt.figure(23)
plt.plot(x,fx)
axes = plt.gca()
axes.set_ylim([-10,1])
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# Plotting 1/f(x)

plt.figure(24)
plt.plot(x,1/fx)
axes = plt.gca()
axes.set_ylim([-5.5,1])
plt.xlabel('x')
plt.ylabel('1/f(x)')
```
